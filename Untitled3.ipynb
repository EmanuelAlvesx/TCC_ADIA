```python
# Importando bibliotecas para geração e visualização de gráfico
import seaborn as sns
import matplotlib.pyplot as plt
pd.options.display.float_format = '{:,.2f}'.format
%matplotlib inline
```

```python
#verificação de outliers
from scipy.stats import zscore

numeric_cols = df1.select_dtypes(include=['float64', 'int64']).columns.tolist()

def detectar_outliers_iqr(serie):
    Q1 = serie.quantile(0.25)
    Q3 = serie.quantile(0.75)
    IQR = Q3 - Q1
    limite_inferior = Q1 - 1.5 * IQR
    limite_superior = Q3 + 1.5 * IQR
    return serie[(serie < limite_inferior) | (serie > limite_superior)]

iqr_outliers = {col: detectar_outliers_iqr(df[col]).count() for col in numeric_cols}

z_scores = df[numeric_cols].apply(zscore)
zscore_outliers = (z_scores.abs() > 3).sum().to_dict()

plt.figure(figsize=(15, 8))
for i, col in enumerate(numeric_cols):
    plt.subplot(1, len(numeric_cols), i + 1)
    sns.boxplot(y=df[col])
    plt.title(f'Boxplot: {col}')
    plt.tight_layout()

plt.show()

print("Outliers por IQR:", iqr_outliers)
print("Outliers por Z-Score:", zscore_outliers)
```

```python
# Geração da base apenas com os atributos, sem normalizacoes.
df3 = df1[['AREA(M2)', 'PRECO(R$)', 'QUARTOS', 'BANHEIROS', 'VAGAS_GARAGEM']]
```

```python
# Início da criação do modelo de machine learning, com separação da vaiável y a ser prevista e as colunas(x) que serão utilizadas
Y = df3['PRECO(R$)']
X = df3.drop('PRECO(R$)', axis=1)
```

```python
# Separação da base em treino e teste
from sklearn.model_selection import train_test_split
x_treino,x_teste,y_treino,y_teste = train_test_split(X,Y)
```

```python
# Utilização do modelo de Regressão Linear
from sklearn.linear_model import LinearRegression

modelo_regressao = LinearRegression()
modelo_regressao.fit(x_treino, y_treino)
y_regressao = modelo_regressao.predict(x_teste)
```

```python
#geração do gráfico de comparação dos valores reais e previstos, a partir do modelo utilizado.
y_real_milhoes = y_teste / 1_000_000
y_previsto_milhoes = y_regressao / 1_000_000
plt.figure(figsize=(8, 6))
plt.scatter(y_real_milhoes, y_previsto_milhoes, alpha=0.6, color='blue', label='Valores Preditos')
min_val = min(y_real_milhoes.min(), y_previsto_milhoes.min())
max_val = max(y_real_milhoes.max(), y_previsto_milhoes.max())
plt.plot([min_val, max_val], [min_val, max_val], 'r--', label='Linha Ideal')  # linha ideal
plt.xlabel("Valores Reais (em milhões)")
plt.ylabel("Valores Preditos (em milhões)")
plt.title("Regressão Linear - Valores Reais vs Preditos")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#avaliando o modelo com o erro quadratico médio
from sklearn.metrics import mean_squared_error
mse_regressao = mean_squared_error(y_teste,y_regressao)
rmse_regressao = np.sqrt(mean_squared_error(y_teste, y_regressao))
print(f"MSE - Regressão Linear: {mse_regressao:.2f}")
print(f"RMSE - Regressão Linear: {rmse_regressao:.2f}")
```

```python
#Avaliando o modelo com R quadrado
from sklearn.metrics import r2_score
r2_regressao = r2_score(y_teste,y_regressao)
print(f"R² - Regresão Linear: {r2_regressao:.4f}")
```

```python
# Avaliando o modelo com erro percentual - MAPE
from sklearn.metrics import mean_absolute_percentage_error
mape_regressao = mean_absolute_percentage_error(y_teste, y_regressao)
print(f"MAPE - Regressão Linear: {mape_regressao:.2%}")
```

```python
# utilizando árvore de decisão
from sklearn.tree import DecisionTreeRegressor
modelo_arvore_decisao = DecisionTreeRegressor()
modelo_arvore_decisao.fit(x_treino,y_treino)
y_arvore_decisao = modelo_arvore_decisao.predict(x_teste)
```

```python
#geração do gráfico de comparação dos valores reais e previstos, a partir do modelo utilizado.
plt.figure(figsize=(8, 6))
plt.scatter(y_teste, y_arvore_decisao, alpha=0.6, label='Valores Preditos')
plt.plot([y_teste.min(), y_teste.max()], [y_teste.min(), y_teste.max()], 'r--', label='Linha Ideal')  # linha ideal
plt.xlabel('Valores Reais')
plt.ylabel('Valores Preditos')
plt.title('Árvore de Decisão - Valores Reais vs Preditos')
plt.legend()
plt.grid()
plt.show()
```

```python
#avaliando o modelo com o erro quadratico médio
mse_arvore = mean_squared_error(y_teste, y_arvore_decisao)
rmse_arvore = np.sqrt(mean_squared_error(y_teste, y_arvore_decisao))
print(f"MSE - Arvore Decisao: {mse_arvore:.2f}")
print(f"RMSE - Arvore Decisao: {rmse_arvore:.2f}")
```

```python
#Avaliando o modelo com R quadrado
r2_arvore = r2_score(y_teste, y_arvore_decisao)
print(f"R² - Arvore Decisao: {r2_arvore:.4f}")
```

```python
# Avaliando o modelo com erro percentual - MAPE
mape_arvore = mean_absolute_percentage_error(y_teste, y_arvore_decisao)
print(f"MAPE - Arvore de Decisao: {mape_arvore:.2%}")
```

```python
#utilizando gradiente boosting
from sklearn.ensemble import GradientBoostingRegressor
modelo_gb = GradientBoostingRegressor()
modelo_gb.fit(x_treino, y_treino)
y_gb = modelo_gb.predict(x_teste)
```
